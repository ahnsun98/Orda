```python
import requests
from bs4 import BeautifulSoup
import json
import re
import pandas as pd
import sqlite3

# json 파일에서 API key 읽어오기
file = open('key.json', 'r').read()
file = json.loads(file)
KEY = file['key']

# API
region = { '서울':'01', '부산':'02', '대구':'03', '인천':'04', '광주':'05', '대전':'06',
         '울산':'07', '경기도':'08', '강원도':'09', '충청북도':'10', '충청남도':'11', '전라북도':'12',
         '전라남도':'13', '경상북도':'14', '경상남도':'15', '제주도':'16' }
season = {'봄':'01', '여름':'02', '가을':'03', '겨울':'04', '봄/여름':'05','봄/가을':'06',
          '봄/겨울':'07', '여름/가을':'08', '여름/겨울':'09', '가을/겨울':'10','사계절':'LL' }

url = 'http://openapi.forest.go.kr/openapi/service/trailInfoService/getforeststoryservice'

# DB
db = sqlite3.connect("C:/Users/User/OneDrive - KT CORPORATION/바탕 화면/AIVLE/강의자료/11. Django/미니프로젝트/Project/openAPIdata/mountain_info.db")
```

```python
res=[]
for key in region.keys(): # 지역별 산 정보 리스트 추출
    params = {'serviceKey' : KEY,
         'mntnNm' : '',
         'mntnHght' : '',
         'mntnAdd' : '',
         'mntnInfoAraCd' : region[key],
         'mntnInfoSsnCd' : '',
         'mntnInfoThmCd' : '' }
    response = requests.get(url, params=params)
    soup = BeautifulSoup(response.text, 'html.parser')

    body = soup.find('items')
    mountain = body.find_all('item')

    for m in mountain: # 필요한 정보만 추출
        name = m.find('mntnnm').string.strip()
        loc = key
        height = m.find('mntninfohght').string.strip()
        overview = m.find('mntninfodscrt').string.strip() #
        routeimg = m.find('rcmmncoursimageseq').string.strip()
        detail = m.find('mntninfodtlinfocont').string.strip() #
        tourism = m.find('crcmrsghtnginfodscrt').string.strip() #
        tourismimg = m.find('crcmrsghtngetcimageseq').string.strip()

        li = []
        for string in [overview, detail, tourism]: # 특정 문자열 전처리
            string = re.sub('[<BR>]+', '. ', string)
            string = re.sub('[^0-9가-힣一-龥 ,.()\n-]', '', string)
            idx = 0
            for s in string:
                if s.isalpha():
                    break
                idx += 1
            string = string[idx:]
            li.append(string)
        overview = li[0]
        detail = li[1]
        tourism = li[2]
        
        res.append([name,loc,height,overview,routeimg,detail,tourism,tourismimg])

# 데이터프레임 생성
mountain_df = pd.DataFrame(res, columns=('name','loc','height','overview','routeimglink','detail',
                                         'tourism','tourismimglink'))

# 데이터프레임 to SQL
mountain_df.to_sql('mountain', db, index_label='id')
```
