```python
seoul = soup.select('mntnnm')
seoul = [mname.string for mname in seoul]
print(seoul)
```

```python
body = soup.find('items')
mountain = body.find_all('item')

res = []
m = mountain[2]
name = m.find('mntnnm').string.strip()
height = m.find('mntninfohght').string.strip()
overview = m.find('mntninfodscrt').string.strip() #
point = m.find('hkngpntdscrt').string.strip() #
routeimg = m.find('rcmmncoursimageseq').string.strip()
detail = m.find('mntninfodtlinfocont').string.strip() #
tourism = m.find('crcmrsghtnginfodscrt').string.strip() #
tourismimg = m.find('crcmrsghtngetcimageseq').string.strip()

import re

li = []
for string in [overview, point, detail, tourism]:
    string = re.sub('[<BR>]+', '. ', string)
    string = re.sub('[^0-9가-힣一-龥 ,.()\n-]', '', string)
    idx = 0
    for s in string:
        if s.isalpha():
            break
        idx += 1
    string = string[idx:]
    li.append(string)
overview = li[0]
point = li[1]
detail = li[2]
tourism = li[3]
    
res.append([name,height,overview,point,routeimg,detail,tourism,tourismimg])

print(res)
```

```python
import pandas as pd

mountain_df = pd.DataFrame(res, columns=('name','height','overview',
                                         'point','routeimglink','detail','tourism','tourismimglink'))
mountain_df.head()
```
